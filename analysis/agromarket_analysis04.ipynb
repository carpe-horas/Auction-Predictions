```python
# 월별 평균 계산
monthly_avg = time_series_data.resample('M').mean()

# 월별 평균 시각화
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg.index, monthly_avg['avg_quantity_kg'], label='월별 평균물량', color='blue')
plt.plot(monthly_avg.index, monthly_avg['avg_unit_price_per_kg'], label='월별 평균단가', color='red')
plt.title('월별 평균물량 및 평균단가')
plt.xlabel('월')
plt.ylabel('값')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 계절별 데이터 그룹화
df_cabbage.loc[:, 'season'] = df_cabbage['date'].apply(lambda x: '봄' if x.month in [3, 4, 5] else
                                                       '여름' if x.month in [6, 7, 8] else
                                                       '가을' if x.month in [9, 10, 11] else '겨울')

# 박스플롯 시각화
plt.figure(figsize=(12, 6))
sns.boxplot(x='season', y='avg_quantity_kg', data=df_cabbage, hue='season', palette='Set3', dodge=False)
plt.title('계절별 평균물량 분포')
plt.ylabel('평균물량 (kg)')
plt.xlabel('계절')
plt.legend(title='계절', loc='upper right')
plt.show()
```

```python
# 자기상관 분석

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# ACF 및 PACF 시각화
plot_acf(time_series_data['avg_quantity_kg'], lags=30)
plt.title('평균물량 ACF')
plt.show()

plot_pacf(time_series_data['avg_quantity_kg'], lags=30)
plt.title('평균물량 PACF')
plt.show()
```

```python
from scipy.stats import f_oneway

# 계절별 단가 데이터 분리
autumn = df_cabbage[df_cabbage['season'] == '가을']['avg_unit_price_per_kg']
winter = df_cabbage[df_cabbage['season'] == '겨울']['avg_unit_price_per_kg']
spring = df_cabbage[df_cabbage['season'] == '봄']['avg_unit_price_per_kg']
summer = df_cabbage[df_cabbage['season'] == '여름']['avg_unit_price_per_kg']

# ANOVA 테스트
f_stat, p_value = f_oneway(autumn, winter, spring, summer)
print(f'F-통계량: {f_stat:.2f}, p-값: {p_value:.4f}')
```

```python
# 귀무가설(계절별 평균단가에 차이가 없다)을 기각
# -> 즉, 계절별 평균단가 차이가 명확히 존재
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# 계절별 평균단가에 대한 Tukey's HSD 테스트
tukey_result = pairwise_tukeyhsd(endog=df_cabbage['avg_unit_price_per_kg'],  # 종속 변수
                                 groups=df_cabbage['season'],               # 그룹(계절)
                                 alpha=0.05)                                # 유의 수준
print(tukey_result)

# 시각화
tukey_result.plot_simultaneous()  # 그룹 간 평균 비교 시각화
plt.title('Tukey HSD Test for Season')
plt.show()
```

```python
# 가을은 모든 계절과 평균 단가에서 유의미한 차이가 존재
# 가을은 김장철과 겹치므로 가격이 높은 경향
```

```python
# 물량 구간 생성
bins = [0, 5000, 10000, 20000, 50000, 100000]
labels = ['0-5,000', '5,000-10,000', '10,000-20,000', '20,000-50,000', '50,000+']
df_cabbage['quantity_bin'] = pd.cut(df_cabbage['avg_quantity_kg'], bins=bins, labels=labels)

# 구간별 평균단가 계산
quantity_price_analysis = df_cabbage.groupby('quantity_bin')['avg_unit_price_per_kg'].mean()

plt.figure(figsize=(10, 6))
quantity_price_analysis.plot(kind='bar', color='skyblue')
plt.title('물량 구간별 평균단가')
plt.xlabel('물량 구간 (kg)')
plt.ylabel('평균단가 (원/kg)')
plt.tight_layout()
plt.show()
```

```python
# 계절별 물량과 평균단가의 관계
plt.figure(figsize=(12, 6))
sns.scatterplot(x='avg_quantity_kg', y='avg_unit_price_per_kg', hue='season', data=df_cabbage, alpha=0.5)
plt.title('계절별 물량과 평균단가의 관계')
plt.xlabel('평균물량 (kg)')
plt.ylabel('평균단가 (원/kg)')
plt.legend(title='계절')
plt.tight_layout()
plt.show()
```

```python
# 계절별 물량과 평균단가 상관계수
seasons = df_cabbage['season'].unique()
for season in seasons:
    seasonal_data = df_cabbage[df_cabbage['season'] == season]
    correlation = seasonal_data['avg_quantity_kg'].corr(seasonal_data['avg_unit_price_per_kg'])
    print(f'{season} 계절 물량과 평균단가 상관계수: {correlation:.2f}')
```

```python
# 고온/저온 구간 생성
bins = [-15, 0, 10, 20, 30, 40]
labels = ['-15~0', '0~10', '10~20', '20~30', '30~40']
df_cabbage['max_temp_bin'] = pd.cut(df_cabbage['max_temp'], bins=bins, labels=labels)
df_cabbage['min_temp_bin'] = pd.cut(df_cabbage['min_temp'], bins=bins, labels=labels)

# 구간별 평균단가와 평균물량 계산
temp_analysis = df_cabbage.groupby('max_temp_bin')[['avg_unit_price_per_kg', 'avg_quantity_kg']].mean()

temp_analysis.plot(kind='bar', figsize=(12, 6), color=['orange', 'skyblue'])
plt.title('고온 구간별 평균단가와 평균물량')
plt.xlabel('최고기온 구간 (°C)')
plt.ylabel('값')
plt.legend(['평균단가 (원/kg)', '평균물량 (kg)'])
plt.tight_layout()
plt.show()
```

```python
# 계절별 고온과 평균단가의 관계
plt.figure(figsize=(12, 6))
sns.boxplot(x='season', y='max_temp', data=df_cabbage, palette='Set3')
plt.title('계절별 최고기온과 평균단가의 관계')
plt.xlabel('계절')
plt.ylabel('최고기온 (°C)')
plt.tight_layout()
plt.show()

# 계절별 저온과 평균물량의 관계
plt.figure(figsize=(12, 6))
sns.boxplot(x='season', y='min_temp', data=df_cabbage, palette='Set2')
plt.title('계절별 최저기온과 평균물량의 관계')
plt.xlabel('계절')
plt.ylabel('최저기온 (°C)')
plt.tight_layout()
plt.show()
```
